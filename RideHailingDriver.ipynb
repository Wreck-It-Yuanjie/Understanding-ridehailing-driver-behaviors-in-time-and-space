```python
def LSq(file_in, file_out):
    Userset = set()
    file_out = open(file_out, 'w')
    flag = 1
    with open(file_in) as fh:
        for line in fh:
            if(flag == 1):
                flag  = flag + 1
                continue
            userID, datetime, email = line.strip("\n").split(",")
            Userset.add(userID.strip("\""))
        print(Userset)
    for user in Userset:
        if(user != "user"):
            file_out.writelines(user + "\n")
    print("Step 1: Output userset done!")
fn = os.path.join(my_path, "RHdriver_table_Source.csv")
LSq(fn, "Userset.csv")
```

### Step 2: Read other files and get the rows containing the userset

```python
def Userset():
    userset = set()
    with open("Userset.csv") as fh:
        for line in fh:
            userID = line.strip("\n").strip(" ")
            userset.add(userID)
        return userset

def get_cmrows(path, fldout):
    files= os.listdir(path)
    userset = Userset()
    try:
        os.stat(os.path.join(path, fldout))
    except:
        os.mkdir(os.path.join(path, fldout))  
    ##print(userset)
    ##print(files)
    for file in files:
        if(file == ".DS_Store"): continue
        fileout = open(os.path.join(path, fldout) + "/" + file, "w")
        if (os.path.isdir(os.path.join(path) + "/" + file)):
            continue
        else:
            with open(os.path.join(path) + "/" + file) as fh:
                flag = 0
                for line in fh:
                    if(flag == 0):
                        flag = 1
                        fileout.writelines(line)
                    user = line.strip("\n").split(",")[0].strip("\"")
                    if(user in userset):
                        print(user)
                        fileout.writelines(line)
    print("Step 2 Success!")
get_cmrows(my_path, "RHdriver_new")
```

### Step 3: For questions that have more than one input per user, only keep the newest one

```python
def kp_nwst(path):
    files= os.listdir(path)
    for file in files:
        if(file == ".DS_Store" or file == "Userset.csv" or file == "RHdriver_table_userlvl_worktime.csv" or file == "RHdriver_table_userlvl_resp.csv"): continue
        df = pd.read_csv(file)
        print(file)
        df = df.sort_values('date').drop_duplicates('user',keep='last')
        df = df.drop(['date'], axis=1)
        df.to_csv(path + "/" + file,index=False)
    print("Step 3 success!")
kp_nwst("RHdriver_new")
```

```python
fileset_resp = ["RHdriver_table_S2_RESP_q1.csv", "RHdriver_table_S2_RESP_q2.csv", "RHdriver_table_S2_RESP_q3.csv", "RHdriver_table_S2_RESP_q4.csv", "RHdriver_table_S2_RESP_q5.csv", "RHdriver_table_S2_RESP_q6.csv"]
expdata_resp = "RHdriver_table_spdata_resp.csv"

fileset_wt = ["RHdriver_table_S2_WTRL_q1.csv", "RHdriver_table_S2_WTRL_q2.csv", "RHdriver_table_S2_WTRL_q3.csv", "RHdriver_table_S2_WTRL_q4.csv", "RHdriver_table_S2_WTRL_q5.csv", "RHdriver_table_S2_WTRL_q6.csv"]
expdata_wt = "RHdriver_table_spdata_worktime.csv"

def MergeExp(choiceset, expdata):
    expd = pd.read_csv(expdata)
    for file in sorted(choiceset):
        print(file)
        df = pd.read_csv(file)
        MergeD = pd.merge(df, expd, how = "left", left_on=['p', "sp"], right_on = ['person', "scenario"])
#         print(MergeD.head())
        MergeD.to_csv(my_path + "/" + "RHdriver_new/" + file, index=False)
    print("Step 4 success!")
# MergeExp(fileset_resp, expdata_resp)
MergeExp(fileset_wt, expdata_wt)
```

### Merge big table

```python
def Merge(path):
    file0 = "RHdriver_table_S0_q1.csv"
    df0 = pd.read_csv(os.path.join(my_path, path) + "/" + file0)
    files= os.listdir(path)
    n = 0
    for file in sorted(files):
        if (file != file0 and file not in [".DS_Store", "Userset.csv", "RHdriver_table_spdata_resp.csv", "RHdriver_table_spdata_worktime.csv"]): 
            print(file)
            df = pd.read_csv(os.path.join(my_path, path) + "/" + file)
            df0 = pd.merge(df0, df, on='user', how = 'outer')
            n = n+1
            ##if(n == 71): break
            df0 = df0.drop_duplicates('user',keep='last')
    df0.to_csv("widedata1.csv",index=False)
    print("Step 5 success!")
Merge("RHdrivertest_new")
```
